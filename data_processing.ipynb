```python
import sparknlp
from pyspark.sql.functions import col
```

```python
spark = sparknlp.start(memory="8g")
spark.sparkContext.setLogLevel("WARN")

df_raw = spark.read.json("arxiv-metadata-oai-snapshot.json")

df_clean = (
    df_raw
    .select(
        "id",
        "title",
        "abstract",
        "categories"
    )
    .filter(
        col("abstract").isNotNull() &
        col("categories").isNotNull()
    )
)

df_clean.write.mode("overwrite").parquet(
    "hdfs:///arxiv/clean"
)

spark.stop()
```
